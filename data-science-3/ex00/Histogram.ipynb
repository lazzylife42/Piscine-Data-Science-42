```python
import pandas as pd
import matplotlib.pyplot as plt
import pathlib

file_path = pathlib.Path().parent.parent/"Test_knight.csv"

knight_DF = pd.read_csv(file_path)
knight_DF.head()
```

```python
num_cols = knight_DF.select_dtypes(include='number').columns
n = len(num_cols)

cols = 5
rows = -(-n // cols)	# arrondi vers le haut

fig, axs = plt.subplots(rows, cols, figsize=(cols * 4, rows * 3))
axs = axs.flatten()

for i, column in enumerate(num_cols):
    knight_DF[column].hist(ax=axs[i], bins=20, color='lightgreen').grid(False)
    axs[i].set_title(column)
    axs[i].set_xlabel('')
    axs[i].set_ylabel('')


fig.tight_layout()
plt.show()
```
